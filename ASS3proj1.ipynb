Now we will calculate the mean and median of 'number of people injured'

```python
mean_injured = np.mean(df['NUMBER OF PERSONS INJURED'])
median_injured = np.median(df['NUMBER OF PERSONS INJURED'])

print("Mean number of persons injured:", mean_injured)
print("Median number of persons injured:" , median_injured)
```

Now we will calculate the standard deviation and correlation between NUMBER OF PERSONS INJURED and NUMBER OF PERSONS KILLED

```python
stdev_injured = np.std(df['NUMBER OF PERSONS INJURED'])

print("The standard deviation of persons injured:" , stdev_injured)

#The 2 Columns we will check for correlation are NUMBER OF PERSONS KILLED AND NUMBER OF PERSONS INJURED
column1 = df['NUMBER OF PERSONS INJURED']
COLUMN2 = df['NUMBER OF PERSONS KILLED']

correlation = np.corrcoef(column1, column2)

corr_coefficient = correlation[0, 1]

print("Correlation coefficient between NUMBER OF PERSONS INJURED AND NUMBER OF PERSONS KILLED:", corr_coefficient)
```
